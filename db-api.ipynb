<h3>Create & Access SQLite database using Python </h3>


#### Create db using sqlite3

```python
import sqlite3
```

```python
# cursor_obj.close()
# conn.close()
conn = sqlite3.connect('INSTRUCTOR.DB')
```

```python
cursor_obj = conn.cursor()
```

#### Create a table in the database

```python
cursor_obj.execute("DROP TABLE IF EXISTS INSTRUCOR")
```

```python
table = """ CREATE TABLE IF NOT EXISTS INSTRUCTOR(ID INTEGER PRIMARY KEY NOT NULL,FNAME VARCHAR(20),LNAME VARCHAR(20),CITY VARCHAR(20),CCODE CHAR(2));"""
cursor_obj.execute(table)
```

```python
cursor_obj.execute("""INSERT INTO INSTRUCTOR VALUES (1,'Pavan','Raj','Bangalore',"IN")""")

```

```python
cursor_obj.execute(""" INSERT INTO INSTRUCTOR VALUES(2,"MOHAN","KUMAR","MAGADI","IN")""")
```

```python
cursor_obj.execute(""" INSERT INTO INSTRUCTOR VALUES(3,"YOGESH","K","HESHRAGATTA","IN"),(4,"SANYUKTA","DAS","WEST BENGAL","IN")""")
```

#### Query data in the table

```python
statement = "select * from INSTRUCTOR"
cursor_obj.execute(statement)

print("all the data")
output = cursor_obj.fetchall()
for row in output:
    print(row)
```

```python
output
```

```python
## fetch few rows from the table

statement = "select * from instructor"
cursor_obj.execute(statement)
for row in cursor_obj.fetchmany(2):
    print(row)
```

```python
# Fetch only FNAME from the table
statement = "select fname from instructor"
cursor_obj.execute(statement)
for row in cursor_obj.fetchall():
    print(row)
```

```python
conn.commit()
```

```python
statement = """update instructor set city = 'BANGALORE' WHERE FNAME = 'SANYUKTA'"""
cursor_obj.execute(statement)
output  = cursor_obj.execute("select * from instructor")
for row in output:
    print(row)
```
